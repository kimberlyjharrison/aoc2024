```python
import pandas as pd
```

```python
data_arr = pd.read_csv('input.txt', header=None)
data = data_arr[0].to_list()
data = [d.split() for d in data]
data = [[int(i) for i in d] for d in data]
```

```python
safe = []

for d in data:
    flag = True
    check_list = []
    for i in range(1, len(d)):
        if flag:
            check_list.append(d[i]-d[i-1])
            if abs(d[i]-d[i-1]) > 3:
                flag=False
    if flag:
        if all(x < 0 for x in check_list) or all(x > 0 for x in check_list):
            safe.append(d)
```

```python
def checkSafe(d):
    flag = True
    check_list = []
    for i in range(1, len(d)):
        if flag:
            check_list.append(d[i]-d[i-1])
            if abs(d[i]-d[i-1]) > 3:
                flag=False
    if flag:
        if all(x < 0 for x in check_list) or all(x > 0 for x in check_list):
            return True
    else:
        return False
    
```

```python
safe = [d for d in data if checkSafe(d)]
unsafe = [d for d in data if not checkSafe(d)]
```

```python
len(safe)
```

```python
for item in unsafe:
    print(item)
    flag = True
    for i in range(len(item)):
        new_list = item[:i] + item[i+1:]
        check = checkSafe(new_list)
        if check:
            safe.append(item)
            break
        
    print("***"*12)
```
